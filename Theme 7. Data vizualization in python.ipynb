## График 5. Распредление спортсменов по годам

```python
sportsmen_born = df[['Player', 'born']].groupby('born').count()
```

```python
plt.plot(sportsmen_born.index, sportsmen_born['Player'])
```

- Какой тип(ы) переменные участвует в визуализации?
int
- Какая цель у данной переменной в визуализации?

- Почему выбран именно такой тип графика? Аргументируйте
график иделано подходит для визуализации исторических данным по годам/месяцам
